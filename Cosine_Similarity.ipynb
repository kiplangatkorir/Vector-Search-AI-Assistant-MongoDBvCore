```python
from sklearn.feature_extraction.text import CountVectorizer
import pandas as pd
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np
```

```python
#Mount Content Drive
from google.colab import drive
drive.mount("/content/drive")
```

```python
#Load the Datasets
df =pd.read_csv("/content/drive/MyDrive/trussdata.csv")
df.head()
```

```python
df_users = pd.DataFrame(df)
```

```python
# Concatenate user profiles into a single text column
df_users['Profile'] = df_users.apply(lambda row: ' '.join(row.drop(['Name', 'Age', 'Gender', 'Location', 'Country', 'JobTitle', 'Industry', 'Skill1', 'Skill2', 'Skill3', 'Hobby1', 'Hobby2']).astype(str)), axis=1)
```

```python
# Vectorize user profiles using CountVectorizer
sparse_matrix = count_vectorizer.fit_transform(df_users['Profile'])
doc_term_matrix = sparse_matrix.todense()
```

```python
# Compute cosine similarity
doc_term_matrix = np.asarray(doc_term_matrix)
cosine_sim = cosine_similarity(doc_term_matrix, doc_term_matrix)
```

```python
print(df_users.index)
print(df_users.columns)
```

```python
cosine_sim = cosine_sim[:df_users.index.shape[0], :df_users.columns.shape[0]]
cosine_sim_df = pd.DataFrame(cosine_sim, index=df_users.index, columns=df_users.columns)
print(cosine_sim_df)
```
